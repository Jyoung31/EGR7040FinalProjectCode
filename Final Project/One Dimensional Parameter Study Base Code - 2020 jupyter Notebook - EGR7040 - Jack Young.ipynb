```python
def fun_eval(xi):
    #from numpy import Matrix
    import math
    c=[0,1,1,5,1]
    yi=c[0]+c[1]*xi[0]**3+c[2]*math.exp(xi[1]*xi[2])-c[3]*xi[1]**2-c[4]*xi[0]*xi[1];
    return yi
```

```python
def study_function(xdata):
    from numpy import append
    sn,nd= xdata.shape
    ydata=[]
    for i in range(0,sn):
        xi=xdata[i,:]
        yi=fun_eval(xi)
        ydata=append(ydata,yi)
    return ydata
```

```python
def param_study1(parameter_to_study):
    from numpy import ones,arange,transpose,array,linspace
    from matplotlib.pyplot import subplots,plot,show,xlabel,ylabel,legend,title,tick_params,grid,ylim,xlim,xticks,yticks
    %matplotlib inline
    fig, ax = subplots()
    sn=10
    xdata=array(ones([sn,3])*.5)
    xs=transpose([array(linspace(0,1,sn))])
    xdata[:,parameter_to_study-1]= xs[:,0]
    ydata=study_function(xdata)
    plot(xs,ydata)
    #title('parameter %i' %parameter_to_study)
    xlabel('Value of parameter '+str(parameter_to_study),fontsize=24)
    ylabel('Objective Function Value',fontsize=24)
    title(' --- parameter '+str(parameter_to_study)+' study ---',fontsize=24)
    fig.set_size_inches(18.5,10.5)
    grid()
```
